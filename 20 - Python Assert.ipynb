# Python Assert

- Python provides the assert statement to check if a given logical expression is true or false. 
- Program execution proceeds if only the expression is true and raises the AssertionError when it is false. 

```python
10>10
```

```python
num = 10
assert num>10
```

```python
try:
    num = int(input("Enter the number "))
    assert num%2 ==0
    print("The number is even")
except AssertionError:
    print("Please Enter a even number")
```

```python
try:
    num = int(input("Enter the number "))
    assert num%2 ==0
    print("The number is even")
except AssertionError:
    print("Please Enter a even number")
```

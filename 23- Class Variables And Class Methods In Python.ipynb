# Class Variables And Class Methods In Python

```python
class Car:
    base_price = 1000000 #class variable
    def __init__(self, window, door, power):
        self.window = window
        self.door = door
        self.power = power
        
    def what_base_price(self):
        print("The base price of car is {}".format(self.base_price))
```

```python
class Car:
    base_price = 1000000 #class variable
    def __init__(self, window, door, power):
        self.window = window
        self.door = door
        self.power = power
        
    def what_base_price(self):
        print("The base price of car is {}".format(self.base_price))
        
    @classmethod
    def revised_base_price(cls, inflation):
        cls.base_price = cls.base_price + cls.base_price * inflation
```

```python
Car.revised_base_price(0.10)
```

```python
Car.base_price
```

```python
car1 = Car(4,5,200)
```

```python
car1.revised_base_price(0.10)
```

```python
car1.base_price
```

# Arguments
- Information can be passed into functions as arguments.

- Arguments are specified after the function name, inside the parentheses. You can add as many arguments as you want, just separate them with a comma.

- The following example has a function with one argument (fname). When the function is called, we pass along a first name, which is used inside the function to print the full name:

# Parameters vs Arguments
The terms parameter and argument can be used for the same thing: information that are passed into a function.

- A parameter is the variable listed inside the parentheses in the function definition.

- An argument is the actual value that is sent to the function when it is called.

```python
def hello(name): #name is Parameter
    print("My name is {}".format(name))
```

```python
hello("Kunal") #Kunal is an argument
```

```python
def hello(name,age):
    print("My name is {} and age is {}".format(name,age))
```

```python
hello("Kunal",24)
```

```python
def hello(name,age):
    return "My name is {} and age is {}".format(name,age)
```

```python
val = hello("Kunal",24)
```

```python
print(val)
```

# Positional-Only Arguments
- You can specify that a function can have ONLY positional arguments.

- To specify positional-only arguments, add , / after the arguments:

```python
def my_f(name, /):
    print("Hello", name)

my_f("Kunal")
```

```python
def my_function(name):
    print("Hello", name)

my_function(name = "Emil")
```

# Keyword-Only Arguments
- To specify that a function can have only keyword arguments, add *, before the arguments:

```python
def my_function(*, name):
    print("Hello", name)

my_function(name = "Emil")
```

# Python *args and **kwargs
- By default, a function must be called with the correct number of arguments.

- However, sometimes you may not know how many arguments that will be passed into your function.

- *args and **kwargs allow functions to accept a unknown number of arguments

```python
def hello(*args, **kwargs):
    print(args)
    print(kwargs)
```

## Arbitrary Arguments - *args
- If you do not know how many arguments will be passed into your function, add a * before the parameter name.

- This way, the function will receive a tuple of arguments and can access the items accordingly:

```python
# Arbitrary Arguments are often shortened to *args in Python documentation.
def child(*kids):
    print("The youngest child is " + kids[2])
```

```python
child("John","Thomas","Harry")
```

# What is *args?
- The *args parameter allows a function to accept any number of positional arguments.

- Inside the function, args becomes a tuple containing all the passed arguments:

```python
def my_function(*args):
    print("Type:", type(args))
    print("First argument:", args[0])
    print("Second argument:", args[1])
    print("All arguments:", args)

my_function("Emil", "Tobias", "Linus")
```

```python
# Using *args with Regular Arguments
def my_function(greeting, *names, **kwargs):
    for name in names:
        print(greeting, name)

my_function("Hello", "Emil", "Tobias", "Linus",10,20)
```

```python
def num(*numbers):
    total = 0
    for sum in numbers:
        total = total+sum
    return total
```

```python
print(num(1,2,3))
print(num(5))
```

```python
def my_function(*numbers):
    if len(numbers) == 0:
        return None
    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num

print(my_function(3, 7, 2, 9, 1))
```

# Arbitrary Keyword Arguments - **kwargs
- If you do not know how many keyword arguments will be passed into your function, add two asterisks ** before the parameter name.

- This way, the function will receive a dictionary of arguments and can access the items accordingly:



```python
def my_function(**kid):
    print("His last name is " + kid["lname"])

my_function(fname = "Tobias", lname = "Refsnes")
```

```python
def student_info(age, *skills, **details):
    print("Age:", age)
    print("Skills:", skills)
    print("Details:", details)

student_info(24, "Python", "SQL", name="Kunal", role="Data Analyst")
```

```python
def my_function(*kid):
    print("His last name is " + kid[1])
my_function("Tobias", "Refsnes")
```

# What is **kwargs?
- The **kwargs** parameter allows a function to accept any number of keyword arguments.

- Inside the function, kwargs becomes a dictionary containing all the keyword arguments:

```python
def my_function(**myvar):
    print("Type:", type(myvar))
    print("Name:", myvar["name"])
    print("Age:", myvar["age"])
    print("All data:", myvar)

my_function(name = "Tobias", age = 30, city = "Bergen")
```

```python
def my_function(username, **details):
    print("Username:", username)
    print("Additional details:")
    for key, value in details.items():
        print(" ", key + ":", value)

my_function("emil123", age = 25, city = "Oslo", hobby = "coding")
```

```python
def my_function(a, b, c):
    return a + b + c

numbers = [1, 2, 3]
result = my_function(*numbers) # Same as: my_function(1, 2, 3)
print(result)
```

```python
def my_function(fname, lname):
    print("Hello", fname, lname)

person = {"fname": "Emil", "lname": "Refsnes"}
my_function(**person) # Same as: my_function(fname="Emil", lname="Refsnes")
```

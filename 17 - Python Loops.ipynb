# Python Loops:-

Python has two permitive loop commands :-

(i) **While** Loops

(ii) **For** Loops

## Pyhton for loops:-
- A for loop is used for iterating over a sequence( that is either a list, tuple, a dict, a set or a string)
- This is less like for keyword in other programming languages, and works more, like an iterator method as found in other object oriented programming language.
- With the 'for' loop we can execute a set of statements. once for each item in a list, tiple, set etc.

### Looping through a string:-


```python
banana = "banana"
for x in banana:
    print(x)
```

### The break statement:-

```python
fruits = ["banana","papaya","mango"]
for x in fruits:
    print(x)
    if x =="papaya":
        break
```

```python
# Exit the loop when x is "banana", but this time the break comes before the print:
fruits = ["banana","papaya","mango"]
for x in fruits:
    if x == "papaya":
        break
    print(x)
```

### The Continue statement:-


```python
fruits = ["banana","papaya","mango"]
for x in fruits:
    if x == "papaya":
        continue
    print(x)
```

### The Range function:-


```python
for x in range(6):
    print(x) 
```

```python
for x in range(2,6):
    print(x)
```

```python
for x in range(2,30,3) #Increment the sequence with 3.
    print(x)
```

### Else in For Loop
- The else keyword in a for loop specifies a block of code to be executed when the loop is finished:

```python
for x in range(6):
    print(x)
else:
    print("Finally Completed")
```

```python
for x in range(6):
    print(x)
    if x == 3:
        break
else:
    print("Finally Completed")
```

### Nested Loops
- A nested loop is a loop inside a loop.

- The "inner loop" will be executed one time for each iteration of the "outer loop":

```python
adj = ["red","big","tasty"]
fruits = ["apple","banana","mango"]
for x in adj:
    for y in fruits:
        print(x,y)
```

### The Pass statement:-
- for loops cannot be empty, but if you for some reason have a for loop with no content, put in the pass statement to avoid getting an error.

```python
for x in [0, 1, 2]:
    pass
```

```python
# Short Hand Statements:
a = 7
b = 6
print("a is greater") if a>b else print("b is greater")
```

```python
lst1 = ["Kunal",1],["Kumar",2],["Marie",250]
for item in lst1:
    print(item[0], "and lolly is",item[1])
```

```python
lst1 = [["Kunal", 1], ["Kumar", 2], ["Marie", 250]]

for name, lollypop in lst1:
    print(name, "and lolly is", lollypop)
```

# While Loops:-

```python
i = 1
while i < 6:
    print(i)
    i += 1
```

### Break Statement:-

```python
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1
```

```python
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```
